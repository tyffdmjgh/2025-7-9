```python
import json

# 定義 Site 類別
class Site:
    def __init__(self,
                 sitename,
                 county,
                 aqi,
                 pollutant,
                 status,
                 pm2_5,
                 pm2_5_avg,
                 latitude,
                 longitude,
                 publishtime):
        self.sitename = sitename
        self.county = county
        self.aqi = aqi
        self.pollutant = pollutant
        self.status = status
        self.pm2_5 = pm2_5
        self.pm2_5_avg = pm2_5_avg
        self.latitude = latitude
        self.longitude = longitude
        self.publishtime = publishtime

# 讀取 JSON 檔案
with open('aqx_p_432.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# 將資料轉成 Site 物件列表
allData = []
for item in data['records']:
    site = Site(
        sitename=item.get('sitename'),
        county=item.get('county'),
        aqi=item.get('aqi'),
        pollutant=item.get('pollutant'),
        status=item.get('status'),
        pm2_5=item.get('pm2.5'),
        pm2_5_avg=item.get('pm2.5_avg'),
        latitude=item.get('latitude'),
        longitude=item.get('longitude'),
        publishtime=item.get('publishtime')
    )
    allData.append(site)

# 印出所有站點資訊
for site in allData:
    print("📍 測站名稱:", site.sitename)
    print("🏙️ 縣市:", site.county)
    print("🌫️ AQI 指數:", site.aqi)
    print("☠️ 主要污染物:", site.pollutant or "無")
    print("🔍 狀態:", site.status)
    print("🧪 PM2.5 現值:", site.pm2_5)
    print("📊 PM2.5 平均值:", site.pm2_5_avg)
    print("📍 緯度/經度:", site.latitude, "/", site.longitude)
    print("🕒 資料時間:", site.publishtime)
    print("=" * 40)
```
